```python
df_d1 = pd.read_csv(d1_path)
df_d2 = pd.read_csv(d2_path) 
df_d3 = pd.read_csv(d3_path)
df_d4 = pd.read_csv(d4_path)

print(df_d1.shape)
print(df_d2.shape)
print(df_d3.shape)
print(df_d4.shape)
```

```python
#df_d1 = df_d1[(df_d1['Rating'] >= 4.0) & (df_d1['Maximum_Installs'] >= 5000)]
#df_d2 = df_d2[(df_d2['Rating'] >= 4.0) & (df_d2['Maximum_Installs'] >= 5000)]
#df_d3 = df_d3[(df_d3['Rating'] >= 4.0) & (df_d3['Maximum_Installs'] >= 5000)]
#df_d4 = df_d4[(df_d4['Rating'] >= 4.0) & (df_d4['Maximum_Installs'] >= 5000)]

df_d1 = df_d1[(df_d1['Rating'] >= 3.5) & (df_d1['Maximum_Installs'] >= 1000)]
df_d2 = df_d2[(df_d2['Rating'] >= 3.5) & (df_d2['Maximum_Installs'] >= 1000)]
df_d3 = df_d3[(df_d3['Rating'] >= 3.5) & (df_d3['Maximum_Installs'] >= 1000)]
df_d4 = df_d4[(df_d4['Rating'] >= 3.5) & (df_d4['Maximum_Installs'] >= 1000)]

print(df_d1.shape)
print(df_d2.shape)
print(df_d3.shape)
print(df_d4.shape)
```

```python
len_d1 = np.array(df_d1['token_length'])
len_d2 = np.array(df_d2['token_length'])
len_d3 = np.array(df_d3['token_length'])
len_d4 = np.array(df_d4['token_length'])

print(len(len_d1))
print(len(len_d2))
print(len(len_d3))
print(len(len_d4))

tot_len = np.concatenate((len_d1, len_d2, len_d3, len_d4), axis=0)
print(len(tot_len))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.figure_factory as ff

sns.set(style='darkgrid')

# Increase the plot size and font size
sns.set(font_scale=1.5)
plt.rcParams["figure.figsize"] = (10,5)

# Truncate any description lengths greater than 512
lengths = [min(l,800) for l in tot_len]

# Plot the distribution of description lengths
sns.distplot(lengths, kde=True, rug=False)


# Alternatively, you might try using a log scale on the x-axis, but this is
# tricky. See here for one approach:
# https:
# plt.xscale('log')

plt.title('Description Lengths')
plt.xlabel('Description Length')
plt.ylabel('# of descriptions') #256
```

```python
lst = []
lst.append(lengths)
group_labels = ['distplot']
fig = ff.create_distplot(lst, group_labels)
fig.show()
```
