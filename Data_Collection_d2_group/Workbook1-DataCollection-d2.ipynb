```python
## Selecting only these categories initially

#----d2-----
## Lifestyle
## Health & Fitness
## Tools
```

```python
df = df[df.apply(lambda x: (x["Category"] in ["Lifestyle", "Health & Fitness","Tools"]) , axis=1)]
```

```python
## keep only the app_name and app_id in the different dataframe
```

```python
df_d2 = pd.DataFrame()
```

```python
df_d2["App_Name"] = df["App_Name"]
df_d2["App_Id"] = df["App_Id"]
df_d2["Category"] = df["Category"]
df_d2["Rating"] = df["Rating"]
df_d2["Maximum_Installs"] = df["Maximum_Installs"]
df_d2["Editors_Choice"] = df["Editors_Choice"]
```

```python
df_d2['Description'] = ''
df_d2['Privacy_Policy'] = ''
```

```python
i = 0
err_list_d2 = pd.DataFrame()
```

```python
for index, row in df_d2[0:345829].iterrows():      #262757 ## 50000 100000 150000 200000 250000 330214
      app_id = row.App_Id
      #print ("Trying...",index,app_id)
      try:
        result = app(app_id)
        df_d2.at[index, 'Description'] = result['description']
        df_d2.at[index, 'Privacy_Policy'] = result['privacyPolicy']
    
      except (Exception, HTTPError, TypeError) as e:
        err_list_d2.at[i,'index'] = index
        err_list_d2.at[i,'app_id'] = app_id
        err_list_d2.at[i,'category'] = row.Category
        i = i + 1
        print ("Error",index,app_id)
```

```python
i = 0
```

```python
perm_set = set()
```

```python
for index, row in df_d2[0:287095].iterrows():
    #299763 ## 50000 100000 150000 200000 250000 286622
    app_id = row.App_Id
    #print ("Trying...",index,app_id)
    try:
        result = permissions(app_id)
        for x in list(result):
            perm_set.add(x)
            #print(x)
            if x != None:
                df_d2.at[index, x] = result[x]

    except TypeError as e:
        remove_list_d2.at[i,'index'] = index
        remove_list_d2.at[i,'app_id'] = app_id
        remove_list_d2.at[i,'category'] = row.Category
        i = i + 1
        print ("Error",index,app_id)
```
