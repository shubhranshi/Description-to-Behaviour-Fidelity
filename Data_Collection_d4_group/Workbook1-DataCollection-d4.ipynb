```python
## keep only the app_name and app_id in the different dataframe
```

```python
df_d4 = pd.DataFrame()
```

```python
df_d4["App_Name"] = df["App_Name"]
df_d4["App_Id"] = df["App_Id"]
df_d4["Category"] = df["Category"]
df_d4["Rating"] = df["Rating"]
df_d4["Maximum_Installs"] = df["Maximum_Installs"]
df_d4["Editors_Choice"] = df["Editors_Choice"]
```

```python
df_d4['Description'] = ''
df_d4['Privacy_Policy'] = ''
```

```python
for index, row in df_d4[0:262398].iterrows():      #262757 ## 50000 100000 150000 200000 250000 330214
      app_id = row.App_Id
      #print ("Trying...",index,app_id)
      try:
        result = app(app_id)
        df_d4.at[index, 'Description'] = result['description']
        df_d4.at[index, 'Privacy_Policy'] = result['privacyPolicy']
    
      except (Exception, HTTPError, TypeError) as e:
        err_list_d4.at[i,'index'] = index
        err_list_d4.at[i,'app_id'] = app_id
        err_list_d4.at[i,'category'] = row.Category
        i = i + 1
        print ("Error",index,app_id)
```

```python
for index, row in df_d4[68832:68834].iterrows():
    #299763 ## 50000 100000 150000 200000 250000 286622
    app_id = row.App_Id
    #print ("Trying...",index,app_id)
    try:
        result = permissions(app_id)
        for x in list(result):
            perm_set.add(x)
            #print(x)
            if x != None:
                df_d4.at[index, x] = result[x]

    except (Exception, HTTPError, TypeError) as e:
        remove_list_d4.at[i,'index'] = index
        remove_list_d4.at[i,'app_id'] = app_id
        remove_list_d4.at[i,'category'] = row.Category
        i = i + 1
        print ("Error",index,app_id)
```
