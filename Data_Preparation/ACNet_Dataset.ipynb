###Preprocessing descriptions

```python
import re
import string
import unicodedata
```

```python
## change text in lowercase
def convert_to_lower(text):
    return text.lower()

## remove numbers
def remove_numbers(text):
    number_pattern = r'\d+'
    without_number = re.sub(pattern=number_pattern, repl=" ", string=text)
    return without_number

## remove URLs
def remove_urls(text):
    rem_url = re.sub(r'http\S+', '', text)
    return rem_url
```

```python
## remove HTML tags

from io import StringIO
from html.parser import HTMLParser

class MLStripper(HTMLParser):
    def __init__(self):
        super().__init__()
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.text = StringIO()
    def handle_data(self, d):
        self.text.write(d)
    def get_data(self):
        return self.text.getvalue()

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()
```

```python
## remove special chars
##  !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~   ##32
##  except !.';?,  ##6

def remove_special_chars(text):
    char_list = ['/', ':', '*', '=', '_', '-', '"', '#', '@', '~', '`', '$', '%', '&', '|', '^', '<', '>', '+']
    char_list_2 = ["\\", "{", "}", "[", "]", "(", ")"]
    for replace_sign in char_list:
        text = text.replace(replace_sign, " ")
    for replace_sign in char_list_2:
        text = text.replace(replace_sign, " ")
    return text


# remove non-ascii
def remove_non_ascii(text):
    text = ''.join(char for char in text if ord(char) < 128)
    return text
```

```python
## remove duplicate whitespaces and newline characters
def remove_extra_white_spaces(text):
    without_sc = " ".join(text.split())
    return without_sc
```

```python
df_ac['Clean_Description'] = ""
```

```python
i = 0
```

```python
for index, row in df_ac[0:1417].iterrows():    #:206048
    print(index)
    #print ("Trying...",index,app_id)
    text = row.description
    
    text = convert_to_lower(text)
    text = remove_numbers(text)
    text = remove_urls(text)
    text = strip_tags(text)
    text = remove_special_chars(text)
    text = remove_non_ascii(text)
    text = remove_extra_white_spaces(text)
    
    df_ac.at[i,'Clean_Description'] = text
    i = i + 1
```
